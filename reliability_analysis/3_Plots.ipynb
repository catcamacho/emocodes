```python
from emocodes import CodeTimeSeries

cts = CodeTimeSeries(sampling_rate=10)
cts.proc_codes_file(codes_file='/Users/catcamacho/Box/CCP/EmoCodes_project/reliability_data/raw/AHKJS1E2_subjective_character_codes_JULIEN_DB.csv', 
                    video_file='/Users/catcamacho/Box/CCP/EmoCodes_project/reliability_data/episodes/AHKJ_S1E2.mp4')
```

```python
import matplotlib.pyplot as plt
from pandas import read_csv

preproc_dir = '/Users/catcamacho/Box/CCP/EmoCodes_project/reliability_data/processed/AHKJ_S1E2/'
obj_codes = read_csv(preproc_dir + 'objective_codes_timeseries_DB_20210316-114734.csv', index_col=0)
obj_codes.index = obj_codes.index/1000
obj_codes['time'] = obj_codes.index
subjnc_codes = read_csv(preproc_dir + 'subjective_noncharacter_codes_timeseries_DB_20210305-100949.csv', index_col=0)
subjnc_codes.index = subjnc_codes.index/1000
subjnc_codes['time'] = subjnc_codes.index
subjchar_codes = read_csv(preproc_dir + 'subjective_character_codes_timeseries_Julien_DB_20210317-105559.csv', index_col=0)
subjchar_codes.index = subjchar_codes.index/1000
subjchar_codes['time'] = subjchar_codes.index
print(obj_codes.columns)
print(subjnc_codes.columns)
print(subjchar_codes.columns)
```

```python
# Objective codes exemplar
plt.figure()
obj_codes.plot(y='has_faces',kind='area',figsize=(6,2), legend=None, title='has_faces', xlabel='Time (seconds)', xlim=(0,obj_codes.index[-1]),linewidth=0.1)
plt.tight_layout()
plt.savefig('obj_codes_exemplar.svg')
plt.show()
plt.close()
```

```python
# Subjective Noncharacter codes exemplar
plt.figure()
subjnc_codes.plot(y='visual_valence_negative',kind='area',figsize=(6,2), legend=None, title='visual_valence_negative', xlabel='Time (seconds)', xlim=(0,subjnc_codes.index[-1]),linewidth=0.1)
plt.tight_layout()
plt.savefig('subjnc_codes_exemplar.svg')
plt.show()
plt.close()
```

```python
# Subjective Character codes exemplar
plt.figure()
subjchar_codes.plot(y='c_happy_face',kind='area',figsize=(6,2), legend=None, title='c_happy_face', xlabel='Time (seconds)', xlim=(0,subjchar_codes.index[-1]),linewidth=0.1)
plt.tight_layout()
plt.savefig('subjchar_codes_exemplar.svg')
plt.show()
plt.close()
```
